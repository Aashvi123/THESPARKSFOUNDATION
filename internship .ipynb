```python
#correlation matrix
plt.subplots(figsize=(15,10))
sns.heatmap(data.corr(),annot=True,linewidths=.5,linecolor='black',fmt='1.1f')
plt.title('correlation heatmap',size=18)
plt.show()
```

We can see that there is a positive correlation between Sales and Profit i.e. Sales increases then Profit increases and vice versa .Similarly,there is a positive correlation between Quantity and Sales.CORRELATION doesn't imply CAUSATION.

HERE,PROFIT IS THE DEPENDENT VARIABLE AND OTHERS ARE INDEPENDENT VARIABLES
